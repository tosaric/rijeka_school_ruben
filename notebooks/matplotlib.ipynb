```python
import matplotlib.style as style 
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 20
#style.available
```

```python
style.use('tableau-colorblind10') #sets the size of the charts
fig, ax = plt.subplots(figsize=(12,10))
x = np.linspace(0,10,100)
for i in range(0,6):
    ax.plot(x, x*i, label=f'{i+1}', lw=3)

ax.legend()
fig.savefig('tableau-colorblind10.png', dpi=400)
```

```python
style.use('seaborn-colorblind') #sets the size of the charts
fig, ax = plt.subplots(figsize=(12,10))
x = np.linspace(0,10,100)
for i in range(0,6):
    ax.plot(x, x*i, label=f'{i+1}', lw=3)

ax.legend()
fig.savefig('seaborn-colorblind.png', dpi=400)
```

```python
mean = [0, 0]
cov = [[100, 0], [0, 100]]

fig, ax = plt.subplots(2, figsize=(12,20))

x, y = np.random.multivariate_normal(mean, cov, 1000000).T
#h = ax.hist2d(x, y, bins=100)
h = ax[1].hist2d(x, y, bins=100, cmap='cividis')
#ax.axis('equal')
fig.colorbar(h[3], ax=ax[1])
h2 = ax[0].hist2d(x, y, bins=100)
fig.colorbar(h2[3], ax=ax[0])
```

```python
ls = ['solid','dashed', 'dashdot', 'dotted' ]
```

```python
style.use('seaborn-colorblind') #sets the size of the charts
fig, ax = plt.subplots(figsize=(12,10))
x = np.linspace(0,10,100)
for i in range(0,4):
    ax.plot(x, x + i, label=f'{ls[i]}', ls=ls[i], lw=3)

ax.legend()
fig.savefig('line_styles.png', dpi=400)
```

```python
ms = ["o", 'v', '^', 's', "P", 'd', "X", "*"]
ms_names = ['circle', 'triangle_down', 'triangle_up', 'square', 'plus', 'diamond', 'cross', 'star']
```

```python
style.use('seaborn-colorblind') #sets the size of the charts
fig, ax = plt.subplots(figsize=(12,10))
x = np.linspace(0,10,10)
for i in range(0, len(ms)):
    #ax.scatter(x, i * np.ones_like(x), label=f'{ms_names[i]}', marker=ms[i], s=200)
    ax.scatter(x, x + i, label=f'{ms_names[i]}', marker=ms[i], s=200)

ax.legend(ncol=2)
fig.savefig('marker_styles.png', dpi=400)
```

```python
hatches = ['-', 'o', 'x', '*', '']
```

```python
style.use('tableau-colorblind10') #sets the size of the charts

fig, ax = plt.subplots(figsize=(12,10))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
langs=['','','','','']
students = [23,17,35,29,12]
patches = ax.pie(students, labels = langs,autopct='%1.0f%%', pctdistance = 1.2)
for i in range(0, len(hatches)):
    patches[0][i].set_hatch(hatches[i])


fig.savefig('pie_chart.png', dpi=400)
```
